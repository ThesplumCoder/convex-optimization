```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
full_img = Image.open("../resources/images/lab4/forest-color-full.png")
print(full_img.size)
plt.imshow(full_img)
```

```python
original = full_img.resize((128, 128))
original.save("../resources/images/lab4/original.jpg")
print(original.size)
plt.imshow(original)
```

```python
low_res = original.resize((64, 64))
low_res.save("../resources/images/lab4/low_res.jpg")
plt.imshow(low_res)
```

```python
high_res = original.convert("L")
high_res.save("../resources/images/lab4/high_res.jpg")
plt.imshow(high_res, cmap = "gray")
```
